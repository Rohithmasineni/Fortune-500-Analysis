```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Importing the cleaned dataset of fortune 500 companies

data = pd.read_csv('Fortune_500.csv')
data
```

### Top 10 companies by Rank

```python
top_10_companies = data.head(10)
top_10_companies
```

### Top 10 companies by revenue

```python

top_10 = data.nlargest(10,'Revenues ($M)')
plt.figure(figsize=(12,5))

# creating a barplot
barplot = sns.barplot(y='Company_Name',x='Revenues ($M)',data = top_10, errorbar = None)

# adding data labels
for bar in barplot.patches:
    barplot.annotate(format(bar.get_width(), ',.0f'),     # adding commas b/w numbers
                     (bar.get_width(), bar.get_y() + bar.get_height() / 2),
                     ha = 'left', va = 'center')
plt.xticks(rotation=45, ha = 'right')
plt.title('Top 10 Companies by Revenue')
# plt.tight_layout()
plt.show()
```

### The barplot illustrates the top 10 companies globally by revenue:

#### 1. Ranking:
The companies are ranked from highest to lowest revenue, with State Grid leading the chart and Industrial & Commercial Bank of China at the bottom.
#### 2. Revenue:
The revenue is displayed on the x-axis, measured in millions of dollars (USD). State Grid boasts the highest revenue, exceeding 5 million USD, while the lowest, Industrial & Commercial Bank of China, still surpasses 2 million USD.
#### 3. Industry Diversity:
The chart showcases a diverse range of industries, including energy, automotive, construction, and banking.
#### 4. Dominance of Chinese Companies:
A significant observation is the strong presence of Chinese companies, occupying half of the top 10 spots.
#### 5. Leaders in Each Sector:
State Grid leads the energy sector, Volkswagen in automotive, and China State Construction Engineering in construction.

### Relation between count of employees and revenue generated

##### Till now we've generated the insights of top most companies and bottom most companies, 
##### Now let's have a look at the rows somewhere in the middle of the data
##### This approach helps us to:-  focus on companies with average employee counts and revenue values, which might represent typical trends

```python
# Range of middle rows
middle_start = len(data) // 4 - 5   # --> Dividing the dataset by 5 and subtracting 5 rows from the result
middle_end = len(data) // 4 + 5    # --> Dividing the dataset by 5 and adding 5 rows to the result

# Extract middle rows
middle_rows = data.iloc[middle_start:middle_end]
middle_rows
```

```python
plt.figure(figsize=(8,4))
plt.scatter(middle_rows['Employees'], middle_rows['Revenue Growth/Decline ($M)'], color = 'blue', alpha = 0.7)
for i in range(len(middle_rows)):
    plt.annotate(
        middle_rows.iloc[i]['Company_Name'],
        (middle_rows.iloc[i]['Employees'], middle_rows.iloc[i]['Revenue Growth/Decline ($M)']),
        fontsize = 7,
        ha = 'right',
        va = 'bottom'
    )

plt.title('Relation between count of employees and revenue generated')
plt.xlabel('Employees', fontsize=12)
plt.ylabel('Revenue Growth/Decline ($M)', fontsize=12)
plt.grid(True)
plt.show()     
    
```

#### The scatter plot shows the relationship between the number of employees a company has and the revenue it generated in 2024.
* There seems to be a positive correlation between the number of employees and revenue generated. This means that, in general, companies with more employees tend to generate higher revenue.
* However, the relationship is not perfectly linear, meaning that other factors also influence revenue besides the number of employees.
* There are a few outliers, such as the company with a large number of employees but relatively low revenue. This suggests that simply increasing the number of employees doesn't guarantee higher revenue.

### Revenue vs. Profits (Profitability)

```python
# Visualization of the relationship between revenues and profits to understand how well companies are converting their revenue into profit.

plt.figure(figsize=(8,4))
plt.scatter(middle_rows['Revenues ($M)'], middle_rows['Profits ($M)'], color = 'red', alpha = 0.7)

for j in range(len(middle_rows)):
    plt.annotate(
    middle_rows.iloc[j]['Company_Name'],
    (middle_rows.iloc[j]['Revenues ($M)'], middle_rows.iloc[j]['Profits ($M)']),
    fontsize = 7,
    ha = 'left',
    va = 'bottom'
    )

plt.title('Relation Between Revenue vs. Profits (Profitability)')
plt.xlabel('Revenues ($M)', fontsize=12)
plt.ylabel('Profits ($M)', fontsize=12)
plt.grid(True)
plt.show() 
    
    
```

#### The scatter plot illustrates the relationship between a company's revenue and its profits, essentially depicting its profitability.
### Key observations:
#### 1) Positive correlation:
The plot shows a generally positive correlation, meaning that as revenue increases, profits tend to increase as well.
#### 2) Outliers:
PepsiCo stands out with exceptionally high profits relative to its revenue, indicating a high-profit margin.
#### 3) Data spread:
The data points are spread out, indicating that the relationship between revenue and profit is not perfectly linear. Several factors beyond revenue influence profitability.

### Top 10 Companies by Profit Margin

```python

# ProfitÂ Margin = (Profits/Revenue) Ã— 100 which visualizes the companies with the highest profit margin.

data['Profit_Margin'] = round((data['Profits ($M)'] / data['Revenues ($M)']) * 100, 2)
```

```python
data
```

```python
top_profit_margin = data.nlargest(10, 'Profit_Margin')
profit_margin = sns.barplot(x='Company_Name', y='Profit_Margin', data=top_profit_margin)
for bar in profit_margin.patches:
    profit_margin.annotate(format(bar.get_height(), ',.0f'),
                           (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                            ha = 'center', va = 'center')
plt.xticks(rotation = 45, ha = 'right')
plt.title('Top 10 Companies by Profit Margin')
plt.show()
```

##### The bar plot shows the top 10 companies by profit margin, with Visa having the highest profit margin at 53% and Apple having the lowest at 25%. 

#### 1) Industry leaders:
Companies like Visa and Nvidia, leading the chart, likely have strong competitive advantages, efficient operations, and high demand for their products or services.
#### 2) Profitability drivers:
The plot highlights the diversity of industries represented, indicating that various factors can contribute to high profit margins, such as brand strength, technological innovation, and market dominance.
#### 3) Investment potential:
Investors might find this plot valuable for identifying companies with strong financial performance and potential for future growth.
#### 4) Comparison across companies:
The plot allows for quick comparisons between companies, making it easy to see which ones are outperforming others in terms of profitability.
